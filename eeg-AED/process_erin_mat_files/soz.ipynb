```python
temp = []
# Iterate through all .mat files in ../../../erinconr/projects/fc_toolbox/results/analysis/intermediate/
for filename in os.listdir(ERIN_DIRECTORY):
    mat_file = scipy.io.loadmat(os.path.join(ERIN_DIRECTORY, filename))
    mat_file = mat_file["summ"][0]
    for index, key in enumerate(mat_file.dtype.names):
        # print(index, key, mat_file[0][index].shape)
        if key == "soz":
            temp.append(index)
soz_index = np.unique(np.array(temp, dtype=int))
assert soz_index.shape[0] == 1
soz_index = soz_index[0]
soz_index
```

```python
for filename in os.listdir(ERIN_DIRECTORY):
    hup_id = int(filename[3:6])
    mat_file = scipy.io.loadmat(os.path.join(ERIN_DIRECTORY, filename))
    mat_file = mat_file["summ"][0][0]
    erin_soz = mat_file[soz_index]
    labels, _, region, lateralization = erin_soz[0][0]
    # Check if labels is not empty
    if labels.size:
        labels = np.concatenate(labels.flatten()).astype(str)
        # Modify each label to insert a '0' between the letter part and the number part
        labels = np.array(["".join([label[:-1], "0", label[-1]]) for label in labels])
    else:
        labels = np.array([])
    print(labels)
    np.save(os.path.join(SOZ_DIRECTORY, f"HUP_{hup_id}.npy"), labels)
```
