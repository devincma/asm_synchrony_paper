```python
temp = []
# Iterate through all .mat files in ../../../erinconr/projects/fc_toolbox/results/analysis/intermediate/
for filename in os.listdir(ERIN_DIRECTORY):
    mat_file = scipy.io.loadmat(f"{ERIN_DIRECTORY}{filename}")
    mat_file = mat_file["summ"][0]
    for index, key in enumerate(mat_file.dtype.names):
        # print(index, key, mat_file[0][index].shape)
        if key == "sz_times":
            temp.append(index)
sz_times_index = np.unique(np.array(temp, dtype=int))
assert sz_times_index.shape[0] == 1
sz_times_index = sz_times_index[0]
```

```python
for filename in os.listdir(ERIN_DIRECTORY):
    patient_hup_id = int(filename[3:6])
    mat_file = scipy.io.loadmat(f"{ERIN_DIRECTORY}{filename}")
    mat_file = mat_file["summ"][0][0]
    erin_sz_times = mat_file[sz_times_index]
    # try:
    #     old_sz_times = np.load(f"{SEIZURES_DIRECTORY}/times/HUP_{patient_hup_id}.npy")
    # except FileNotFoundError:
    #     print(f"FileNotFoundError: HUP {patient_hup_id}")
    #     continue
    # # Assert sz_times and old_sz_times are close
    # assert np.allclose(erin_sz_times, old_sz_times, atol=300)

    print(f"Saving HUP {patient_hup_id} {erin_sz_times.shape}")

    # Save sz_times
    np.save(f"{SEIZURES_DIRECTORY}/source_mat/HUP_{patient_hup_id}.npy", erin_sz_times)
```
