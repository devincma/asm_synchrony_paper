```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Spike train parameters
time_window = 2  # minutes
num_spikes = 10

# For evenly spaced out spikes
even_spike_times = np.linspace(0, time_window, num_spikes)

# For randomly spaced out spikes (sorted to ensure increasing time)
random_spike_times = np.sort(np.random.rand(num_spikes) * time_window)

# Raster Plots
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

axs[0].vlines(even_spike_times, 0, 1, color="black")
axs[0].set_title("Evenly spaced spikes")
axs[0].set_yticks([])
axs[0].spines["top"].set_visible(False)
axs[0].spines["right"].set_visible(False)
axs[0].spines["left"].set_visible(False)

axs[1].vlines(random_spike_times, 0, 1, color="black")
axs[1].set_title("Randomly spaced spikes")
axs[1].set_yticks([])
axs[1].spines["top"].set_visible(False)
axs[1].spines["right"].set_visible(False)
axs[1].spines["left"].set_visible(False)
axs[1].set_xlabel("Time (minutes)")

plt.tight_layout()
plt.show()

# Bar Plot for Spike Rates
spike_rate_even = num_spikes / time_window
spike_rate_random = num_spikes / time_window

plt.figure(figsize=(3, 6))
plt.bar(
    ["Evenly spaced", "Randomly spaced"],
    [spike_rate_even, spike_rate_random],
    color=["blue", "green"],
)
plt.ylabel("Spike Rate (spikes/min)")
plt.title("Spike Rate")
plt.show()

# Bar Plot for Fano Factor
# Calculate inter-spike intervals (ISIs)
ISI_even = np.diff(even_spike_times)
ISI_random = np.diff(random_spike_times)

# Calculate Fano factor (variance / mean)
fano_factor_even = np.var(ISI_even) / np.mean(ISI_even)
fano_factor_random = np.var(ISI_random) / np.mean(ISI_random)

plt.figure(figsize=(3, 6))
plt.bar(
    ["Evenly spaced", "Randomly spaced"],
    [fano_factor_even, fano_factor_random],
    color=["blue", "green"],
)
plt.ylabel("Fano Factor")
plt.title("Fano Factor of Inter Spike Intervals")
plt.show()
```
