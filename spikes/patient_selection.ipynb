# Select Patients for Interictal Spike Detection

```python
import numpy as np
import pandas as pd
```

Dr. Erin Conrad has validated the spike detector for patients. Patients whose PPV (car) value is lower than 70% should be discarded.

```python
erin_df = pd.read_excel("../../Data/erin_spike_detector_validation.xlsx")
# Drop the ID column
erin_df = erin_df.drop(columns=["ID"])
# Drop rows that have NaNs
erin_df = erin_df.dropna()
# Drop rows where PPV (car) is lower than 0.70
erin_df = erin_df[erin_df["PPV (car)"] >= 0.70]
# Extract number from name
erin_df["hup_id"] = erin_df["name"].str.extract(r"(\d+)")
# Make hup_id column an integer
erin_df["hup_id"] = erin_df["hup_id"].astype(int)
```
