```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
SYNCHRONY_60_100_DIRECTORY = "../../Data/synchrony/hourly/60_100"
SYNCHRONY_1_70_DIRECTORY = "../../Data/synchrony/hourly/1_70"
SYNCHRONY_broadband_DIRECTORY = "../../Data/synchrony/hourly/broadband"
```

```python
for hup_id in hup_ids:
    arrays = {
        "synchrony_60_100": np.load(f"{SYNCHRONY_60_100_DIRECTORY}/HUP_{hup_id}.npy"),
        "synchrony_1_70": np.load(f"{SYNCHRONY_1_70_DIRECTORY}/HUP_{hup_id}.npy"),
        "synchrony_broadband": np.load(
            f"{SYNCHRONY_broadband_DIRECTORY}/HUP_{hup_id}.npy"
        ),
    }

    plt.figure(figsize=(15, 5))

    for idx, (name, array) in enumerate(arrays.items()):
        plt.subplot(1, 3, idx + 1)  # 1 row, 3 columns of subplots

        # Plot distribution
        sns.histplot(array[~np.isnan(array)], bins=40, kde=True)
        plt.title(name)
        plt.xlabel("Synchrony Value")
        plt.xlim(0, 1)

        # Remove y-label for the second and third columns
        if idx in [1, 2]:
            plt.ylabel("")

    plt.suptitle(f"HUP {hup_id}", fontsize=14)
    plt.tight_layout()
    plt.show()
```
